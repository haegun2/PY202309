```python
import random #랜덤값 사용을 위해 작성

#숫자를 랜덤으로 범위지정된 곳에서 가져오는 함수를 작성
def get_number():
    return random.randrange(1,46)
#로또 숫자를 생성함. 리스트를 만들고, 중복이 아니라면 리스트에 추가
def set_lotto_numbers():
    lotto = []
    num = 0
    while True:
        num = get_number()
    
        if lotto.count(num) == 0:
            lotto.append(num)
        #lotto 리스트의 갯수가 여섯개가 된다면 그만 가져옴
        if len(lotto) >= 6:
            break
    return lotto #로또 리스트를 반환

#내 로또 값과, 컴퓨터 랜덤 로또 값의 리스트를 더해서 set 해줌으로 중복을 제거함. 12-합한 리스트 를 하면 중복값을 알 수 있음
def match_numbers(user_inputs, lotto_numbers):
    hap_numbers =  set(user_inputs +lotto_numbers)
    same_numbers = 12 - len(hap_numbers)
    return same_numbers #중복값을 반환

#반환된 중복값을 기반으로 맞춘 숫자별 상이 뭔지 출력
def check_prize(same_numbers):
    if same_numbers == 6:
        print("1st prize")
    elif same_numbers == 5:
        print("2nd prize")
    elif same_numbers == 4:
        print("3rd prize")
    elif same_numbers == 3:
        print("4th prize")
    else:
        print("no prize :(")

        
num = 0
print("** Start Lotto **")
print("1) Set the lotto numbers")
lotto_numbers = set_lotto_numbers() #컴퓨터 로또값을 함수를 통해 불러옴

print("2) Set the user inputs") #유저 로또값을 기입받음
user_inputs = []
while True:
    num = int(input("Enter %d-number: "% (len(user_inputs)+1)))
    if num <1 or num > 46: #범위를 벗어나는 값이면 다시 값을 받음
        print("No balid number")
        continue
    if user_inputs.count(num) == 0: #받은 값을 검토해봐도 없으면 리스트에 추가함
        user_inputs. append(num)
    else:
        print("Same number") #같은 값이 들어오면 같은값이라고 출력하고 다시 값을 받음
        continue
    if len(user_inputs) >= 6 :# 유저 복권값의 숫자가 6개가 되면 그만 받음
        break
    
print("3) Match the numbers") #유저랑 컴퓨터의 로또값의 중복값의 갯수를 알기위해 함수 호출
same_numbers = match_numbers(user_inputs, lotto_numbers)

print("4) Check the prize")# 상을 알기 위해 중복값을 함수로 보냄
check_prize(same_numbers)

print("5) Sort the lists")# 유저의 값과 컴퓨터의 값을 오름차순으로 변경
lotto_numbers.sort()
user_inputs.sort()

print("===Summary===")
print("Lotto numbers: ", *lotto_numbers) #괄호를 땐 값으로 출력
print("User numbers: ", *user_inputs)
```
